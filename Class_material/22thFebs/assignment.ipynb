```python
import requests
from bs4 import BeautifulSoup

url = "https://www.youtube.com/@PW-Foundation/videos"
response = requests.get(url)

soup = BeautifulSoup(response.text, 'html.parser')

# Find all video containers
video_containers = soup.find_all('div', {'class': 'text-wrapper style-scope ytd-video-renderer'})

# Extract URLs for the first five videos
for i in range(5):
    video_container = video_containers[i]
    video_link = video_container.find('a', {'class': 'yt-simple-endpoint style-scope ytd-video-renderer'})
    video_url = "https://www.youtube.com" + video_link['href']
    print(video_url)
```
