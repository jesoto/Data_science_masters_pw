1. What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Matplotlib is a Python data visualization library that allows you to create static, animated, and interactive visualizations in Python. It is one of the most widely used libraries for data visualization in Python.

Matplotlib is used for creating a wide variety of visualizations, including line plots, scatter plots, bar plots, histograms, 3D plots, and many more. It is widely used in the scientific community, data science, and machine learning applications for visualizing data and results.

The Pyplot module of Matplotlib is a sublibrary that provides a simple interface for creating a wide range of plots with a minimum amount of code. Some of the most common plots that can be created using the Pyplot module of Matplotlib include:

- Line plot: used to show trends or changes over time.
- Scatter plot: used to show the relationship between two variables.
- Bar plot: used to compare the values of different categories or groups.
- Histogram: used to show the distribution of a single variable.
- Pie chart: used to show the relative sizes of different categories or groups as parts of a whole.


There are many other types of plots that can be created using Matplotlib, and the library provides a lot of flexibility for customizing the visualizations.

2. What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
- A scatter plot is a type of plot in which data points are plotted on a two-dimensional plane, with one variable on the x-axis and the other variable on the y-axis. Each data point is represented as a dot, and the position of the dot on the graph represents the values of the two variables for that data point

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt

plt.scatter(x,y)
plt.title('First scatter plot')
plt.xlabel('X AXIS')
plt.ylabel('ylabel')
```

3. Why is the subplot() function used? Draw four line plots using the subplot() function.

```python
x1 = np.array([0, 1, 2, 3, 4, 5])  
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([50, 20, 40, 20, 60, 70])
x3= np.array([0, 1, 2, 3, 4, 5])  
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])  
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
fig, axs = plt.subplots(4, 1, figsize=(10, 10))
axs[0].plot(x1, y1)
axs[0].set_title('Line 1')
axs[1].plot(x2, y2)
axs[1].set_title('Line 2')
axs[2].plot(x3, y3)
axs[2].set_title('Line 3')
axs[3].plot(x4, y4)
axs[3].set_title('Line 4')
```
